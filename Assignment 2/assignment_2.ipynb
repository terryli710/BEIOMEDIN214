```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import LeaveOneOut
```

```python
leukemia = pd.read_csv('leukemia.csv', header=0)
leukemia
```

```python
x = leukemia.iloc[:,:-1]
y = leukemia.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
loo = LeaveOneOut()
loo.get_n_splits(x)
```

```python
preds = []
true = []
for train_index, test_index in loo.split(x):
    x_train, x_test = x.iloc[train_index,:], x.iloc[test_index,:]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    knn = KNeighborsClassifier()
    knn.fit(x_train, y_train)
    preds.append(knn.predict(x_test)[0])
    true.append(y_test.iloc[0])
```

```python
# q1
import numpy as np
np.mean([preds[i]==true[i] for i in range(len(preds))])
```

```python
# true positive for all
print(np.sum([preds[i]==true[i]=='ALL' for i in range(len(preds))]))
# false positive for all
print(np.sum([preds[i]=='ALL' and true[i]=='AML' for i in range(len(preds))]))
```

```python
# true positive for aml
print(np.sum([preds[i]==true[i]=='AML' for i in range(len(preds))]))
# false positive for aml
print(np.sum([preds[i]=='AML' and true[i]=='ALL' for i in range(len(preds))]))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
## NB classifier
preds2 = []
true2 = []
for train_index, test_index in loo.split(x):
    x_train, x_test = x.iloc[train_index,:], x.iloc[test_index,:]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    gnb = GaussianNB()
    gnb.fit(x_train, y_train)
    preds2.append(gnb.predict(x_test)[0])
    true2.append(y_test.iloc[0])
```

```python
# true positive for all
print(np.sum([preds2[i]==true2[i]=='ALL' for i in range(len(preds2))]))
# false positive for all
print(np.sum([preds2[i]=='ALL' and true[i]=='AML' for i in range(len(preds2))]))
```

```python
# true positive for aml
print(np.sum([preds2[i]==true2[i]=='AML' for i in range(len(preds2))]))
# false positive for aml
print(np.sum([preds2[i]=='AML' and true2[i]=='ALL' for i in range(len(preds2))]))
```
